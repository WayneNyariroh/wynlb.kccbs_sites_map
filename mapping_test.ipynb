```python
import pandas as pd
import folium
from folium.plugins import FastMarkerCluster
```

```python
sites_data = pd.read_excel("processed_data/cleaned_data.xlsx", usecols = {'mfl_code', 'facility_name', '2023Q1', '2023Q2', '2023Q3', '2023Q4',
                                                                             'county', 'sub_county','lat', 'lon', 'region'})
```

```python
sites_data.dtypes
```

```python
sites_data['lat'] = pd.to_numeric(sites_data['lat']).apply(float)
sites_data['lon'] = pd.to_numeric(sites_data['lon']).apply(float)
```

```python
map = folium.Map(location=[ -1.286389, 36.817223], zoom_start=7, min_zoom=3, max_zoom=11)

site_coordinates = (sites_data['lat'], sites_data['lon'])
folium.TileLayer('cartodbpositron').add_to(map)
folium.Marker(location=site_coordinates, icon="icons/txregime.png").add_to(map)
```
